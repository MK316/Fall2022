# gTTS (Google Text-to-Speech)

```python
#@markdown ðŸš© {gTTS} package installation and import
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio
```

```python
text = input()
def tts(mytext):
  text_to_say = mytext
  language_to_choose = "en" #@param ["en", "ko", "fr", 'es']
  language = language_to_choose
  gtts_object = gTTS(text = text_to_say,
                     lang = language,
                    slow = False) 
  
  gtts_object.save("mytext.wav")
  return Audio("mytext.wav")

text_to_say = text
intro_text = "Okay. I'll say words and you try to identify the vowel in each word"

text_to_say = intro_text + 'Ready?' + text_to_say
tts(text_to_say)
```
